# Valores faltantes

```python
dados.info()
```

Não tem um único jeito de tratamento de valores nulos ou valores faltantes. Quais as formas que podemos tratar esses dados?

1. Deletar todas as linhas que tenho valores faltantes; Desvantagem: se a nossa tabela tem muitos valores nulos, provavelmente a gente vai perder muita informação deletando tudo. Fica bem chato se tivermos poucos dados para analisar.

2. Substituir os valores faltantes pela média ou pela mediana. O bom é que a gente não vai perder nenhum dado, já que não deletamos nada e funciona bem com uma base de dados pequena ou homogênea. Para valores categóricos, usamos a moda (o valor que se repete mais)
Desvantagem: não leva em consideração a covariância dos fatores. Tem fatores que variam em conjunto, por ex, quanto mais chove, aumenta a quantidade de pessoas que vai ao cinema, mas se chove MUITO, diminui essa quantidade. Só substituir pela média pode fazer com que a relação entre essas duas variáveis não sejam correspondidas nos valores substituidos.

3. Criar modelo de aprendizado de máquina para resolver o problema. Pode dar ótimos resultados e vai levar em consideração a covariância dos dados. A desvantagem é que é muito mais trabalhoso.

# Intervalo de confiança e distribuição amostral
Intervalo de confiança: faixa de valores onde a média da população provavelmente está (tipo um palpite que tem uma boa chance de estar certo)
É como dar uma margem de erro para o palpite da sua amostra

Distribuição amostral: é como ter uma espécie de repetição dos experimentos imaginários

erro padrão

```python
salarios = dados['SALARIO']
```

```python
salarios
```

```python
media_amostral = np.mean(salarios)
media_amostral
```

```python
desvio_amostral = np.std(salarios)
desvio_amostral
```

```python
nivel_confianca = 0.95
```

```python
tamanho_amostral = len(salarios)
tamanho_amostral
```

```python
from scipy import stats
```

```python
erro_padrao = stats.sem(salarios)
erro_padrao
```

```python
intervalo_confianca = stats.t.interval(nivel_confianca, tamanho_amostral - 1, loc=media_amostral, scale=erro_padrao)
intervalo_confianca
```

# Featuring engineering

om, essa engenharia de recursos é o processo de selecionar, criando e transformando

Imagine que você está fazendo um bolo. Um bom exemplo da engenharia

```python
def preencher_nivel(gestor, nivel):
  if gestor == 1:
    return "Pessoa gestora"
  else:
    return nivel
```

Eu achava que quanto maior a idade, maior o salário, né?
Mas pelos nossos dados, dá pra perceber que a idade não é o fator mais importante para fazer o
salário aumentar. E isso nos leva até a novas perguntas. Qual será que é o fator mais importante
para se obter um aumento salarial?

```python
dados.columns
```

```python
from scipy.stats import chi2_contingency
```

```python
def cramer_coeficiente(coluna1, coluna2):
  tabela_cruzada = np.array(pd.crosstab(coluna1, coluna2))
  chi2 = chi2_contingency(tabela_cruzada)[0] #retorna apenas o primeiro valor
  soma = np.sum(tabela_cruzada)
  mini = min(tabela_cruzada.shape) -1
  cramer = np.sqrt(chi2/(soma*mini))
  return cramer
```

```python
cramer_coeficiente(dados['COR/RACA/ETNIA'], dados['NIVEL DE ENSINO'])
```

```python
cramer_coeficiente(dados['GENERO'], dados['NIVEL DE ENSINO'])
```
